```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
colleges = pd.read_csv("ForbesAmericasTopColleges2019.csv")
```

```python
colleges.info()
```

```python
ar = np.array(colleges['Acceptance Rate'])
```

```python
ar = ar[:50]
```

```python
ar
```

```python
rank = np.array(colleges['Rank'])
```

```python
rank = rank[:50]
```

```python
rank
```

```python
T = np.arctan2(rank, ar)
```

```python
plt.scatter(rank, ar, s=75, c=T, alpha=0.5)
```

```python
m,b = np.polyfit(rank, ar, 1)
```

```python
m,b
```

```python
plt.plot(rank, rank*m+b)
plt.scatter(rank, ar, s=75, c=T, alpha=0.5)
```

```python
curve = np.polyfit(rank, ar, 2)
```

```python
curve
```

```python
poly = np.poly1d(curve)
```

```python
poly
```

```python
xx = []
yy = []
for i in range(50):
    xx.append(i+1)
    yy.append(poly(i+1))
```

```python
plt.plot(xx,yy)
plt.scatter(rank, ar, s=75, c=T, alpha=0.5)
plt.xlabel("Rank")
plt.ylabel("Acceptance Rate")
plt.title("Colleges Rank vs Acceptance Rate in the US")
```
